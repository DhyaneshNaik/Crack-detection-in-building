```python
import os
import numpy
import matplotlib.pyplot as plt
%matplotlib inline

import keras
from keras.preprocessing.image import ImageDataGenerator
```

```python
dataset_dir = 'data/'
```

```python
data_generator = ImageDataGenerator(rescale=1./255)
```

```python
# gather images for classes from directory in atched of 4
image_generator = data_generator.flow_from_directory(dataset_dir,
                                                   batch_size=4,
                                                   class_mode='categorical',
                                                   seed = 24)
```

```python
# get next batch
first_image = image_generator.next()
first_image
```

```python
# get next batch of arrays of images
first_batch_images = image_generator.next()[0]
first_batch_images
```

```python
# get next batch of labels if images
first_batch_labels = image_generator.next()[1]
first_batch_labels
```

```python
first_batch = image_generator.next()[0]
first_batch
```

```python
fig, axs = plt.subplots(nrows=2,ncols=2,figsize = (20,10))

ind = 0
for ax1 in axs:
    for ax2 in ax1:
        images_data = first_batch[ind]
        ax2.imshow(images_data)
        ind += 1
        
fig.suptitle("First batch of concrete images")
plt.show()
```

```python
second_batch = image_generator.next()[0]
second_batch
```

```python

fig, axs = plt.subplots(nrows=2,ncols=2,figsize = (20,10))

ind = 0
for ax1 in axs:
    for ax2 in ax1:
        images_data = second_batch[ind]
        ax2.imshow(images_data)
        ind += 1
        
fig.suptitle("Second batch of concrete images")
plt.show()
```
